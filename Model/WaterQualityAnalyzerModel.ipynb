```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("water_potability.csv")
df
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
mean_ph= df.ph.mean()
mean_ph
```

```python
mean_sulfate= df.Sulfate.mean()
mean_sulfate
```

```python
mean_tri= df.Trihalomethanes.mean()
mean_tri
```

```python
df.ph=df.ph.fillna(mean_ph)
df.Sulfate=df.Sulfate.fillna(mean_sulfate)
df.Trihalomethanes=df.Trihalomethanes.fillna(mean_tri)
```

```python
df.isna().sum()
```

```python
df.shape
```

```python
#Check Potability count
df[df.Potability==0].shape
```

```python
df[df.Potability==1].shape
```

```python
#Safe PH range is 6.5 to 8.5, we remove other points 
df=df[~((df.ph<6.5)&(df.Potability==1))]
df=df[~((df.ph>8.5)&(df.Potability==1))]
```

```python
df
```

```python
#Hardness of safe drinking water is in range 0-300, we remove any points above that
df[(df.Hardness>300)&(df.Potability==1)]
```

```python
df=df[~((df.Hardness>300)&(df.Potability==1))]
```

```python
df.shape
```

```python
#Tubidity range of safe drinking water is between 0-5, we remove the other points
df[((df.Turbidity>5))&(df.Potability==1)]
```

```python
df=df[~((df.Turbidity>5)&(df.Potability==1))]
```

```python
df.shape
```

```python
df[df.Potability==1]
```

```python
df[df.Potability==0]
```

```python
X= df.drop(['Potability'],axis='columns')
X
```

```python
Y= df.Potability
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import ShuffleSplit
```

```python
def find_best_model_using_gridsearch(X,Y):
    algos={
        'logistic_regression':{
            'model': LogisticRegression(),
            'params': {}
        },
        'svm':{
            'model': SVC(),
            'params':{'C':[1,5,10],'kernel':['linear','poly']}
        },
        'random_forest':{
            'model': RandomForestClassifier(),
            'params':{'n_estimators':[20,30,50]}
        }
    }
    scores=[]
    cv= ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs= GridSearchCV(config['model'],config['params'],cv=cv,return_train_score=False)
        gs.fit(X,Y)
        scores.append({
            'model':algo_name,
            'best_params':gs.best_params_,
            'best_score':gs.best_score_
        })
    return pd.DataFrame(scores,columns=['model','best_params','best_score'])
find_best_model_using_gridsearch(X,Y)
```

```python
rf=RandomForestClassifier(n_estimators=50)
rf.fit(X_train,Y_train)
```

```python
rf.score(X_test,Y_test)
```

```python
rf.predict(X_test)
```
